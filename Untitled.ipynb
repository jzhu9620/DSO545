```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```


# 1. Density Plot

```python
data=pd.read_csv('mpg(1).csv')
data.head()
```

```python
plt.figure()
sns.kdeplot(data.loc[data.cyl==4,'cty'],
            shade=True,color='green',
           label='Cyl=4',
           alpha=0.4)
sns.kdeplot(data.loc[data.cyl==5,'cty'],
            shade=True,color='pink',
           label='Cyl=5',
           alpha=0.4)
sns.kdeplot(data.loc[data.cyl==6,'cty'],
            shade=True,color='blue',
           label='Cyl=6',
           alpha=0.4)
sns.kdeplot(data.loc[data.cyl==8,'cty'],
            shade=True,color='orange',
           label='Cyl=8',
           alpha=0.4)
plt.ylim(0.0,0.5)
plt.xlim(5,40)

plt.show()
```

```python
#3 Create 2x3 grid
plt.figure()
plt.subplots_adjust(hspace=0.5,wspace=0.6)
for i in range(1,7):
    plt.subplot(2,3,i)
    plt.text(0.5,0.5,str((2,1,i)),ha='center') #ha is horizontal allignment

plt.show()
```

```python
#3 create a 2x3 graph with sin and cos plots

plt.figure()
fig,ax=plt.subplots(2,3,sharex='col',sharey='row')
x=np.linspace(0,50)
ax[0,0].plot(np.sin(x))
ax[0,2].plot(np.cos(x),color='red')
#ax[0,0] means the plot with loc=row0,column0

plt.show()
```

```python
#4
fig=plt.figure()
grid=plt.GridSpec(2,3,wspace=0.5,hspace=0.5)

fig.add_subplot(grid[0,0])
fig.add_subplot(grid[0,1:3])
fig.add_subplot(grid[1,0:2])
fig.add_subplot(grid[1,2])
plt.show()
```

# 5.Marginal boxplot

```python
fig=plt.figure(figsize=(10,5))
grid=plt.GridSpec(4,4,wspace=0.4,hspace=0.4)

ax_main=fig.add_subplot(grid[0:3,0:3])
ax_right=fig.add_subplot(grid[0:3,3])
ax_bottom=fig.add_subplot(grid[3,:3])

#plot the main graph (scatter plot)
ax_main.scatter(data.displ,data.hwy)
ax_main.set(xticks=[2,4,6],yticks=[20,30,40],ylabel='hwy')

#plot the right graph(bar plot)
sns.boxplot(data.hwy,ax=ax_right,orient='v')
ax_right.set(xticks=[],yticks=[],ylabel='')

#plot the bottom graph(bar plot)
sns.boxplot(data.displ,ax=ax_bottom)
ax_bottom.set(xticks=[],yticks=[],xlabel='')

#set title, xlabel,ylabel
ax_main.set(title='Scatterplot with Histograms\ndispl vs hwy')

plt.show()
```

```python
#6.

data=pd.read_csv('economics.csv')
data.head()
```

```python
fig=plt.figure(figsize=(12,5))

plt.plot(data.date,data.psavert,color='tab:red') #another red color
ax=plt.gca()

#set the main y-axis (left)
ax.tick_params(axis='y',labelcolor='tab:red')
ax.set_ylabel('Personal Savings Rate',color='tab:red',fontsize=14)

#set the secondary y-axis (right)
ax_right=ax.twinx() ### create a twin y axis that share the same x-axis
ax_right.plot(data.date,data.unemploy,color='tab:blue')
ax_right.tick_params(axis='y',labelcolor='tab:blue')
ax_right.set_ylabel("#Unemployed(1000's)",color='tab:blue',fontsize=14)

##set the x-axis
ax.set_xticks(np.arange(0,len(data.date),60))
ax.set_xlabel('Year',fontsize=14)

plt.title('Personal Savings Rate vs. Unemployed',fontsize=20)
plt.show()
```
